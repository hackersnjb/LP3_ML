```python
df.pickup_datetime=pd.to_datetime(df.pickup_datetime,utc=0)
```

```python
df['dropoff_longitude']=df['dropoff_longitude'].fillna(value=df['dropoff_longitude'].mean())
df['dropoff_latitude']=df['dropoff_latitude'].fillna(value=df['dropoff_latitude'].mean())
```

```python
df=df.assign(hour=df.pickup_datetime.dt.hour,
         year=df.pickup_datetime.dt.year,
         month=df.pickup_datetime.dt.month,
         dayofweek=df.pickup_datetime.dt.dayofweek)
```

```python
def remove_outliers(df1,col):
    Q1=df1[col].quantile(0.25)
    Q3=df1[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    df[col]=np.clip(df1[col],lower,upper)
    return df1[col]
```

```python
def treat_outliers(df1,col_list):
    for c in col_list:
        df[c]=remove_outliers(df1,c)
    return df
```

```python
df=treat_outliers(df,df.columns)
```

```python
df.plot(kind="box",subplots=True,layout=(7,2),figsize=(17,13))
```

```python
import haversine as hs
travel_dist=[]
for c in range(len(df.pickup_latitude)):
    lati1,longi1,lati2,longi2=[df['pickup_latitude'][c],df['pickup_longitude'][c],df['dropoff_latitude'][c],df['dropoff_longitude'][c]]
    loc1=(lati1,longi1)
    loc2=(lati2,longi2)
    x=hs.haversine(loc1,loc2)
    travel_dist.append(x)
```

```python
df['travel_dist_km']=travel_dist
```

```python
df.shape
```

```python
df=df.loc[(df.travel_dist_km>=1) | (df.travel_dist_km<=130)]
```
